### import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 11.1 Visualization methods

```python
# data load

## Iris
from sklearn.datasets import load_iris
iris = load_iris()
ir = pd.DataFrame(data=iris.data, columns=iris.feature_names)
ir_species = pd.Series(iris.target)
ir_species = ir_species.map({0:"s", 1:"c", 2:"v"})

## Crabs data
lcrabs = pd.read_csv("../data/lcrabs.csv")
lcrabs_grp = list(np.repeat(["B","b","O","o"], [50]*4, axis = 0))
```

### 1) Principal Component analysis

#### Principal Component for the iris data.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

std = StandardScaler().fit_transform(np.log(ir))
pca = PCA()
ir_pca = pca.fit_transform(std)
```

```python
plt.bar(['Comp.1', 'Comp.2', 'Comp.3', 'Comp.4'], pca.explained_variance_)
plt.ylabel('variances', fontsize = 18)
plt.show()
```

```python
pca = PCA(n_components=2)
ir_pca = pca.fit_transform(std)
pca_columns=['Comp1','Comp2']
ir_pca_df = pd.DataFrame(-ir_pca, columns=pca_columns)

markers=['$s$', '$c$', '$v$']
for i, marker in enumerate(markers):
    x_axis_data = ir_pca_df.loc[iris.target==i,:]['Comp1']
    y_axis_data = ir_pca_df.loc[iris.target==i,:]['Comp2']
    
    plt.scatter(x_axis_data, y_axis_data, marker=marker,label=iris.target_names[i])

plt.legend(loc='upper right')
plt.xlabel('first principal component')
plt.ylabel('second principal component')
plt.show()
```

#### Principal Component for the crabs data.

```python
std = StandardScaler().fit_transform(lcrabs)
pca = PCA(n_components=2)
lcrabs_pca = pca.fit_transform(std)
lcrabs_pca_df = pd.DataFrame(data = -lcrabs_pca, columns = ['Comp1', 'Comp2'])
lcrabs_pca_df = pd.concat([lcrabs_pca_df, pd.DataFrame(data = lcrabs_grp, columns = ['group'])], axis = 1)

fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('first principal component', fontsize = 15)
ax.set_ylabel('second principal component', fontsize = 15)

targets = ["B","b","O","o"]
colors = ['blue', 'skyblue', 'orange', 'yellow']
for target, color in zip(targets, colors):
    indicesToKeep = lcrabs_pca_df['group'] == target
    ax.scatter(lcrabs_pca_df.loc[indicesToKeep, 'Comp1']
               , lcrabs_pca_df.loc[indicesToKeep, 'Comp2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

### 3) Distance methods

```python
from sklearn import manifold

mds = manifold.MDS(2, dissimilarity='euclidean', metric = True)
coords = mds.fit_transform(ir)
x, y = coords[:,0], coords[:,1]

fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot(1, 1, 1)

targets = ['setosa', 'versicolor', 'virginica']
colors = ['r', 'g', 'b']

ax.scatter(x, y, c = list(np.repeat(['r', 'g', 'b'], [50]*3, axis = 0)), s = 15)
ax.legend(targets)
ax.grid()
```
